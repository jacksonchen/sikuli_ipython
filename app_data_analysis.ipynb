```python
import pandas as pd
from IPython.display import HTML
```

```python
df_permissions = pd.read_csv('/Users/Jackson/dev/ipython-test/data_analysis/additional_info.csv')
df_permissions.rename(columns={'Package Name': 'package', 'version_code':'verc'}, inplace=True)
df_permissions.sort(columns=['package', 'verc'], inplace=True)
df_permissions.set_index(['package','verc'], inplace=True)
df_permissions
```

```python
def get_diff(dataframe, permission_column_name, permission_result_column_name, rating_column_name, ratingresult_column_name):
    # Get the first index name (package name)
    first_index = dataframe.index.names[0]
    df_result = pd.DataFrame(columns=[first_index, 'number_of_versions', permission_result_column_name, ratingresult_column_name])
    df_result.set_index(first_index, inplace =True)
    for gp in dataframe.index.get_level_values(first_index).unique():
        tmp_pmdf = dataframe.loc[gp, permission_column_name]
        tmp_rtngdf = dataframe.loc[gp, rating_column_name]
        rtng = 0
        perm = 0
        if len(tmp_pmdf) == 1:
            perm = 0
            rtng = 0
        else:
            for r in tmp_pmdf:
                perm = r - perm
            for g in tmp_rtngdf:
                rtng = g - rtng
        df_result.ix[gp, permission_result_column_name] = perm
        df_result.ix[gp, ratingresult_column_name] = rtng
        df_result.ix[gp, 'number_of_versions'] = len(tmp_pmdf)
    return df_result
```

```python
df_permissions_changes = get_diff(df_permissions, 'total_permissions', 'permission_changes', 'rating', 'rating_changes')
df_permissions_changes.sort(columns=['permission_changes', 'rating_changes'], inplace=True)
df_permissions_changes
```

```python
ax = df_permissions_changes.plot(x="permission_changes", y="rating_changes", figsize=(12,8), title="Relationship between average rating change versus permissions change for Android apps")
ax.set_xlabel("Changes in permissions")
ax.set_ylabel("Average rating change")
```
