```python
import pandas as pd
from IPython.display import HTML
```

```python
df_permissions = pd.read_csv('/Users/Jackson/dev/ipython-test/data_analysis/additional_info.csv')
df_permissions.rename(columns={'Package Name': 'package', 'version_code':'verc'}, inplace=True)
df_permissions.sort(columns=['package', 'verc'], inplace=True)
df_permissions.set_index(['package','verc'], inplace=True)
df_permissions
```

```python
def get_diff(dataframe, permission_column_name, permission_result_column_name):
    # Get the first index name (package name)
    first_index = dataframe.index.names[0]
    df_result_plus = pd.DataFrame(columns=[first_index, 'number_of_versions', permission_result_column_name])
    df_result_minus = pd.DataFrame(columns=[first_index, 'number_of_versions', permission_result_column_name])
    df_result_plus.set_index(first_index, inplace =True)
    df_result_minus.set_index(first_index, inplace =True)
    for gp in dataframe.index.get_level_values(first_index).unique():
        tmp_pmdf = dataframe.loc[gp, permission_column_name]
        perm = 0
        if len(tmp_pmdf) == 1:
            perm = 0
        else:
            for r in tmp_pmdf:
                perm = r - perm
        if perm > 0:
            df_result_plus.ix[gp, 'number_of_versions'] = len(tmp_pmdf)
            df_result_plus.ix[gp, permission_result_column_name] = perm
        elif perm < 0:
            df_result_minus.ix[gp, permission_result_column_name] = perm
            df_result_minus.ix[gp, 'number_of_versions'] = len(tmp_pmdf)
    return df_result_plus, df_result_minus
```

```python
df_permissions_changes_plus, df_permissions_changes_minus = get_diff(df_permissions, 'total_permissions', 'permission_changes')
df_permissions_changes_minus.sort(columns=['permission_changes'], inplace=True)
df_permissions_changes_plus.sort(columns=['permission_changes'], inplace=True)
df_permissions_changes_plus
```

```python
f = open('added.html','wb')
f.write(df_permissions_changes_plus.to_html())
f.close()
e = open('deleted.html','wb')
e.write(df_permissions_changes_minus.to_html())
e.close()
```
